### Proyecto de simulación de inteligencia artificial  utiliza  diccionarios,con if, else, elif  while ,funciones,tipo Menu
### 18. Diseñador: Portafolio digital, software de diseño, estilo artístico, clientes activos, tarifa por proyecto y tipo de hardware.

```python
# --- INICIALIZACIÓN ---
# El nombre del diccionario empieza vacío [6]
diseñadores = {}

# --- FUNCIONES DEL SISTEMA [7, 8] ---

def reporte_inteligencia():
    """Analiza la densidad de datos en el diccionario principal [9, 10]."""
    # Se evalúa la cantidad de llaves en el diccionario
    cantidad = len(diseñadores)
    print("\n" + "="*60)
    print("   MONITOR DE IA: ESTADO DEL NÚCLEO 'VISION-DESIGN'   ")
    print("="*60)

    if cantidad == 0:
        print("SISTEMA: VACÍO. La base de datos no contiene registros.")
    elif cantidad < 3:
        print("SISTEMA: BAJA DENSIDAD. IA en modo de aprendizaje inicial.")
    elif cantidad < 5:
        print("SISTEMA: ÓPTIMO. Diversidad de perfiles detectada.")
    else:
        print("SISTEMA: ALTA CARGA. Diccionario cerca del límite de proceso.")
    print("="*60)

def registrar_diseñador():
    """Crea una entrada única en el diccionario principal [11]."""
    print("\n>>> CAPTURA DE DATOS PARA NUEVO REGISTRO")

    # El nombre servirá como LLAVE ÚNICA (Key) para el diccionario principal
    nombre = input("Nombre del Diseñador (Llave): ")

    # Creamos el sub-diccionario de valores (Values) [6]
    detalles = {}
    detalles["portafolio"] = input("URL Portafolio digital: ")
    detalles["software"] = input("Software de diseño: ")
    detalles["estilo"] = input("Estilo artístico: ")
    detalles["clientes"] = input("Clientes activos: ")
    detalles["tarifa"] = input("Tarifa por proyecto (USD): ")
    detalles["hardware"] = input("Tipo de hardware: ")

    # Asignamos el sub-diccionario al diccionario principal usando el nombre como llave [11]
    diseñadores[nombre] = detalles
    print(f"\nIA: Registro de '{nombre}' procesado y almacenado.")

def mostrar_tabla():
    """Recorre el diccionario para presentar los datos alineados [12, 13]."""
    if len(diseñadores) == 0:
        print("\n>>> IA ERROR: No hay datos para generar la tabla.")
    else:
        # Encabezado con formato de alineación de cadenas [12]
        print("\n" + "-"*130)
        print(f"{'DISEÑADOR':<20} | {'SOFTWARE':<20} | {'ESTILO':<20} | {'CLIENTES':<10} | {'TARIFA':<10} | {'HARDWARE':<20}")
        print("-" * 130)

        # Ciclo FOR para recorrer pares CLAVE-VALOR del diccionario [13, 14]
        for nombre, info in diseñadores.items():
            # Accedemos a los valores del sub-diccionario 'info' mediante sus llaves [11]
            print(f"{nombre:<20} | {info['software']:<20} | {info['estilo']:<20} | {info['clientes']:<10} | ${info['tarifa']:<9} | {info['hardware']:<20}")
        print("-" * 130)

# --- BUCLE PRINCIPAL (MENÚ INTERACTIVO) [15] ---

while True:
    reporte_inteligencia()

    print("\nCENTRAL DE COMANDOS IA:")
    print("1. Registrar Diseñador")
    print("2. Visualizar Base de Datos (Tabla)")
    print("3. Purgar Diccionario (Borrar todo)")
    print("4. Apagar Sistema")

    opcion = input("\nIngrese instrucción numérica: ")

    # Lógica de decisión [9, 16]
    if opcion == "1":
        registrar_diseñador()
    elif opcion == "2":
        mostrar_tabla()
    elif opcion == "3":
        # Se limpia el diccionario para que vuelva a estar vacío [17]
        diseñadores.clear()
        print("\n>>> IA: Memoria purgada exitosamente.")
    elif opcion == "4":
        print("\n>>> IA: Desconectando... Núcleo fuera de línea.")
        break # Rompe el ciclo para finalizar [15, 18]
    else:
        print("\n>>> IA ERROR: Comando no reconocido.")
```
